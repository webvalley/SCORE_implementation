```python
#importing csv file
data_frame = pd.read_excel("./Data/PLIC_Milano1234.xlsx")
```

```python
for i, el in enumerate(data_frame.columns):
    print (i, el)
```

```python
#function to control smoking column values as 'ex', 'other'.... and convert them to 'si
def check_smoking_column(column):
        for i,el in enumerate(data_frame[column]):
                if (not(el.startswith('si', 0,len(data_frame[column][i])) or el.startswith('no', 0,len(data_frame[column][i])))):
                    data_frame[column][i]='si'
                    print(i, "fatto")
```

```python
#function for the smoking column and changing every 'si' in 1 and every 'no' in 0
def bool_smoking_column(column):
        for i,el in enumerate(data_frame[column]):
            if (el=='si'):
                data_frame[column][i]=1
                print(i, "fatto")
            elif (el=='no'):
                data_frame[column][i]=0
                print(i, "fatto")
```
